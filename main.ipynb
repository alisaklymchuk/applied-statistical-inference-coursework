```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
ist_full_df = pd.read_csv("IST-data.CSV")
ist_full_df.head()
print(list(ist_full_df.columns))
```

```python
ist_df = ist_full_df[["SEX", "AGE", "RSBP", "RCONSC", "FDEAD", "RXASP"]]
ist_df = ist_df.rename(columns={
    "RSBP": "Systolic BP",
    "RCONSC": "Conscious Level",
    "RXASP": "Aspirin Allocated",
    "FDEAD": "Dead"
}).dropna()
ist_df["SEX"] = ist_df["SEX"].map({"M": 0, "F": 1})
ist_df["Conscious Level"] = ist_df["Conscious Level"].map({"D": 0, "F": 1, "U": 2})
ist_df = ist_df[ist_df["Dead"].isin(["Y", "N"])]
ist_df["Dead"] = ist_df["Dead"].map({"N": 0, "Y": 1}).astype(int)
ist_df["Aspirin Allocated"] = ist_df["Aspirin Allocated"].map({"N": 0, "Y": 1})
```

```python
ist_df["Dead"].unique()
```

```python
sex_counts = ist_df["SEX"].value_counts().reset_index()
sex_counts.columns = ["SEX", "count"]  # rename for clarity

fig_sex = px.bar(
    sex_counts,
    x="SEX",       # categories: M, F
    y="count",     # how many of each
    text="count",  # show numbers on top of bars (optional)
    title="Number of men and women in IST dataset"
)

fig_sex.show()
```
